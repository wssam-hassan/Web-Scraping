```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

### Code

```python
def scrape_today_matches(yallakora_url = 'https://www.yallakora.com/Match-Center/%D9%85%D8%B1%D9%83%D8%B2-%D8%A7%D9%84%D9%85%D8%A8%D8%A7%D8%B1%D9%8A%D8%A7%D8%AA#nav-menu'):
    """
    Extracts the mathced info. from Yallakora website.
    Args:
        url (str): The URL to Yallakora website's table of matches
    Return:
        dataframe of table of matches
    """
    
        # Get the website info at the moment
    yallakora = requests.get(yallakora_url)

        # Beautiful Souping the website request for easy scraping
    soup = BeautifulSoup(yallakora.text, features="html.parser")
    
        # Extracting matches info.
    first_teams = [ soup.find_all(class_='teamA')[i].find(class_='teamName').text.replace('\n', '').strip()
               for i in range(len(soup.find_all(class_='teamA')))]

    second_teams = [ soup.find_all(class_='teamB')[i].find(class_='teamName').text.replace('\n', '').strip()
                   for i in range(len(soup.find_all(class_='teamA')))]

    match_result = [ soup.find_all(class_='matchResult')[i].text.replace('\n', '').strip().replace(' ', ' - ')
                   for i in range(len(soup.find_all(class_='teamA')))]

    match_time = [ soup.find_all(class_='matchTime')[i].text.replace('\n', '').strip()
                   for i in range(len(soup.find_all(class_='teamA')))]
    
        # Save data in Dataframe and show it.    
    df = pd.DataFrame({ 'First Team'   : first_teams,
                        'Match Result' : match_result,
                        'Second Team'  : second_teams,
                        'Match Time'   : match_time})
    
    return df
```

### Test the code

```python
scrape_today_matches()
```
